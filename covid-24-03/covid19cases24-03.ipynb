```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

%matplotlib inline
```

```python
data = pd.read_csv('brazil_covid19.csv')
data = data.drop('hour', axis = 1)
today = data[data['date'] == '2020-03-24']
most_cases = today.sort_values( by = 'cases', ascending = False )[:5]
```

```python
historic = data.groupby(['date']).sum()
```

```python
plt.plot(historic.index ,historic['cases'])
plt.xticks(rotation=45)
plt.rcParams["figure.figsize"] = [22,12]
plt.title('Cases Covid-19 in Brazil',fontsize = 35)
plt.ylabel('Number of cases', fontsize=24)
plt.savefig('cases24-03.pdf')
mpl.rc('xtick', labelsize=13)
mpl.rc('ytick', labelsize=20)
```

```python
plt.plot(historic.index ,historic['deaths'])
plt.xticks(rotation=45)
plt.rcParams["figure.figsize"] = [22,12]
plt.title(' Deaths from covid-19 in Brazil',fontsize = 35)
plt.ylabel('Number of deaths', fontsize=24)
plt.savefig('deaths24-03.pdf')
mpl.rc('xtick', labelsize=13)
mpl.rc('ytick', labelsize=20)
```

```python
plt.plot(historic.index ,historic['suspects'])
plt.xticks(rotation=45)
plt.rcParams["figure.figsize"] = [22,12]
plt.title(' Suspected cases of covid-19 in brazil',fontsize = 35)
plt.ylabel('Number of suspects', fontsize=24)
plt.savefig('suspects-covid-24-03.pdf')
mpl.rc('xtick', labelsize=13)
mpl.rc('ytick', labelsize=20)
```
